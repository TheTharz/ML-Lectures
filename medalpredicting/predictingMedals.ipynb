Creating the model and finding co-efficients

```python
import pandas as pd
import numpy as np
```

```python
teams = pd.read_csv('teams.csv')
```

```python
teams
```

```python
X = teams[["athletes","prev_medals"]].copy()
y = teams[["medals"]].copy()
```

```python
X
```

```python
y
```

```python
X["intercept"] = 1
```

```python
X = X[["intercept","athletes","prev_medals"]]
```

```python
X
```

```python
X_T = X.T
X_T
```

```python
B = np.linalg.inv(X_T@X)@X_T@y
```

```python
B
```

```python
B.index = X.columns
```

```python
B
```

```python
predictions = X@B
predictions
```
